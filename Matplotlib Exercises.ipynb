```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 5, 3]
plt.scatter(x, y)
plt.title("Simple Scatter Plot")
plt.xlabel("X-values")
plt.ylabel("Y-values")
plt.show()
```

```python
import matplotlib.pyplot as plt
categories = ['A', 'B', 'C']
values = [5, 7, 3]
plt.bar(categories, values)
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
data = np.random.randn(1000)
plt.hist(data, bins=30, edgecolor='black')
plt.title("Histogram of Normally Distributed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 5, 3]
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].plot(x, y, color='blue', marker='o')
axes[0].set_title("Line Plot")
axes[0].set_xlabel("X-values")
axes[0].set_ylabel("Y-values")
axes[1].bar(x, y, color='orange')
axes[1].set_title("Bar Chart")
axes[1].set_xlabel("Categories")
axes[1].set_ylabel("Values")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
values = [1, 2, 3, 4]
plt.plot(values, 'r--')
plt.title("Red Dashed Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
y1 = [1, 2, 3]
y2 = [3, 2, 1]
x = [1, 2, 3] 
plt.plot(x, y1, label="Line 1 (y1 = 1,2,3)", color='blue')
plt.plot(x, y2, label="Line 2 (y2 = 3,2,1)", color='red')
plt.title("Two Lines with Legend")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [0, 2, 4, 6, 8, 10]
y = [0, 3, -2, 4, -1, 5]
plt.plot(x, y, marker='o', color='purple')
plt.xlim(0, 10)
plt.ylim(-5, 5)
plt.title("Plot with Custom Axis Limits")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
